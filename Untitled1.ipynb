```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA



```

```python
# Load dataset
file_path = "ifood_df.csv"
df = pd.read_csv(file_path)
```

```python
# Data Exploration
print("Dataset Info:")
df.info()
print("\nMissing Values:")
print(df.isnull().sum())
print("\nSummary Statistics:")
print(df.describe())

```

```python
# Select numerical features for clustering
features = [
    "Income", "MntWines", "MntFruits", "MntMeatProducts", "MntFishProducts",
    "MntSweetProducts", "MntGoldProds", "NumDealsPurchases", "NumWebPurchases",
    "NumCatalogPurchases", "NumStorePurchases", "NumWebVisitsMonth", "Recency"
]
```

```python
# Standardize data
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df[features])
```

```python
# Determine optimal K using Elbow Method
inertia = []
k_range = range(2, 10)
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(df_scaled)
    inertia.append(kmeans.inertia_)
```

```python
# Plot Elbow Curve
plt.figure(figsize=(8, 5))
plt.plot(k_range, inertia, marker="o", linestyle="-", color="b")


```

```python
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Inertia")
plt.title("Elbow Method for Optimal K")
plt.grid(True)
plt.show()
```

```python
# Apply K-Means with K=4
kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
df["Cluster"] = kmeans.fit_predict(df_scaled)

```

```python
# Reduce dimensions with PCA for visualization
pca = PCA(n_components=2)
df_pca = pca.fit_transform(df_scaled)
df["PCA1"], df["PCA2"] = df_pca[:, 0], df_pca[:, 1]

```

```python
# Visualize clusters (sampled to avoid memory issues)
df_sample = df.sample(n=500, random_state=42)
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df_sample["PCA1"], y=df_sample["PCA2"], hue=df_sample["Cluster"], palette="viridis", alpha=0.7)
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.title("Customer Segments Visualization (Sampled)")
plt.legend(title="Cluster")
plt.grid(True)
plt.show()

```

```python
# Analyze cluster characteristics
cluster_means = df.groupby("Cluster")[features].mean()

```

```python
# Plot mean values across clusters
cluster_means.T.plot(kind="bar", figsize=(12, 6), colormap="viridis")
plt.title("Cluster-wise Average Feature Values")
plt.ylabel("Average Value")
plt.xlabel("Features")
plt.xticks(rotation=45, ha="right")
plt.legend(title="Cluster")
plt.grid(axis="y")
plt.show()

```
